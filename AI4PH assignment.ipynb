```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read COVID death data

df = pd.read_csv(r'C:\Users\janyn\OneDrive\Desktop\AI4PH\ai4ph-2024\covid_deaths.csv')
```

```python
df.head()
```

```python
# This code plots the trend of total deaths and deaths related to COVID-19 over time.

df['date'] = pd.to_datetime(df['date'])

# Plot
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['deaths_total'], label='Total Deaths', marker='o')
plt.plot(df['date'], df['death_covid'], label='COVID-19 Deaths', marker='o')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.title('Total Deaths vs. COVID-19 Deaths over Time')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
# This code calculates summary statistics of total deaths and deaths due to COVID-19 over year month. 

# Convert 'date' column to datetime type
df['date'] = pd.to_datetime(df['date'])

# Extract year and month from the 'date' column
df['year_month'] = df['date'].dt.to_period('M')

# Group by year-month and calculate mean and standard deviation for 'deaths_total' and 'death_covid'
summary_stats = df.groupby('year_month').agg({'deaths_total': ['mean', 'std'], 'death_covid': ['mean', 'std']})

# Print summary statistics
print("Summary Statistics by Year-Month:")
print(summary_stats)
```

# Findings:

I created a visualization that plots the trend of total deaths and deaths related to COVID-19 over time and created two summary statistics for total deaths and COVID-19 related deaths (mean and standard deviation).

The visualization shows that total deaths and COVID-19 deaths vary over time, with fluctuations observed from month to month and year to year. There appears to be some seasonal variations for COVID-19 deaths across different months, as seen in peaks in January 2021 and January 2022.

The mean COVID-19 deaths peaked in April 2020, indicating the highest impact of the pandemic during that period.

There is an overall decrease in deaths, as shown with both total deaths and COVID-19 deaths show decreasing, especially evident after the peak in April 2020.

The data suggests, in the most recent months (December 2023-January 2024), a relatively low number of both total deaths and COVID-19 deaths compared to previous periods. This could indicate a potential decline in disease severity or successful control measures such as the COVID-19 vaccine.
